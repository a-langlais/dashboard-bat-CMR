```python
import pandas as pd
import pyproj
df = pd.read_csv(r'data/data_antenna_2016_2023.csv', low_memory = False)
def convert_l93_to_latlon(x, y):
    transformer = pyproj.Transformer.from_crs("epsg:2154", "epsg:4326", always_xy=True)
    lon, lat = transformer.transform(x, y)
    return lat, lon

df['LONG_WGS'], df['LAT_WGS'] = convert_l93_to_latlon(df['LONG_L93'], df['LAT_L93'])
```

```python
df.to_csv(r'data/data_antenna_2016_2023.csv', index = False)
```

```python
# Fonction pour supprimer les guillemets et remplacer la virgule par un point
def clean_decimal(time_str):
    return time_str.replace('"', '').replace(',', '.')

# Nettoyage des données dans la colonne 'HEURE'
df['HEURE'] = df['HEURE'].apply(clean_decimal)

# Fonction pour convertir les heures décimales en HH:MM:SS
def convert_decimal_to_time(val):
    try:
        # Conversion en float et vérification si la valeur est une fraction de journée
        float_val = float(val)
        if 0 <= float_val <= 1:
            total_seconds = float_val * 86400  # Total de secondes dans une journée
            hours = int(total_seconds // 3600)
            minutes = int((total_seconds % 3600) // 60)
            seconds = int(total_seconds % 60)
            return f'{hours:02}:{minutes:02}:{seconds:02}'
        else:
            return val
    except ValueError:
        return val

# Application de la conversion
df['HEURE'] = df['HEURE'].apply(convert_decimal_to_time)
```

```python
df = df.rename(columns={'LONG_WGS': 'LAT_WGS_temp', 'LAT_WGS': 'LONG_WGS'})
df = df.rename(columns={'LAT_WGS_temp': 'LAT_WGS'})
```

```python
occurences = df['NUM_PIT'].value_counts()
occurences = occurences.sort_values(ascending = False)
print(occurences)
```

```python
df.sort_values(by=['NUM_PIT', 'LIEU_DIT', 'DATE'], inplace=True)

# Identifier les interruptions (chaque fois que la différence de jours entre les enregistrements est >1)
df['Prev_DATE'] = df.groupby(['NUM_PIT', 'LIEU_DIT'])['DATE'].shift(1)
df['New_Visit'] = (df['DATE'] - df['Prev_DATE']).dt.days > 1
df['Visit_ID'] = df.groupby(['NUM_PIT', 'LIEU_DIT'])['New_Visit'].cumsum()

# Calculer min/max dates pour chaque visite
result = df.groupby(['NUM_PIT', 'LIEU_DIT', 'Visit_ID']).agg(
    Start=pd.NamedAgg(column='DATE', aggfunc='min'),
    Finish=pd.NamedAgg(column='DATE', aggfunc='max')
).reset_index()

# Résultat des paires min/max pour chaque individu à chaque site
print(result)
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

def create_trajectories_map(df):
    if df.empty:
        fig = go.Figure(go.Scattermapbox())
        fig.update_layout(mapbox_style="open-street-map", mapbox_zoom=6, mapbox_center={"lat": 46.493889, "lon": 2.602778})
        return fig

    # Définir les centres de la carte basés sur les moyennes des latitudes et longitudes
    center_lat, center_lon = df['LAT_WGS'].mean(), df['LONG_WGS'].mean()

    fig = go.Figure()

    # Extraction des liaisons uniques entre sites
    unique_connections = df.sort_values('DATE').drop_duplicates(subset=['LAT_WGS', 'LONG_WGS'], keep='first')

    # Tracer les liaisons
    last_point = None
    for _, row in unique_connections.iterrows():
        if last_point is not None:
            fig.add_trace(go.Scattermapbox(
                lat=[last_point['LAT_WGS'], row['LAT_WGS']],
                lon=[last_point['LONG_WGS'], row['LONG_WGS']],
                mode='lines',
                line=dict(width=2),
                showlegend=False
            ))
        last_point = row

    # Ajouter les marqueurs pour les sites, avec noms dans la légende
    sites = df.drop_duplicates(subset=['LAT_WGS', 'LONG_WGS', 'LIEU_DIT'])
    for _, site in sites.iterrows():
        fig.add_trace(go.Scattermapbox(
            lat=[site['LAT_WGS']],
            lon=[site['LONG_WGS']],
            mode='markers+text',
            marker=go.scattermapbox.Marker(size=9, color='blue'),
            text=site['LIEU_DIT'],
            textposition="bottom right",
            name=site['LIEU_DIT'],  # Utiliser le lieu dit comme nom dans la légende
            showlegend = False
        ))

    fig.update_layout(
        mapbox_style="open-street-map",
        mapbox_zoom=6,
        mapbox_center={"lat": center_lat, "lon": center_lon},
        height = 1000,
        width = 1000
    )
    return fig

%time create_trajectories_map(df)
```
